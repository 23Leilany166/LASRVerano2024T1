```python
def calcular_clases(datos):
    clases = sorted(set(datos))  # Obtener clases Ãºnicas ordenadas
    return clases

def asignar_indices(datos):
    clases_unicas = calcular_clases(datos)
    clase_a_indice = {clase: idx for idx, clase in enumerate(clases_unicas)}
    indices_clases = [clase_a_indice[elem] for elem in datos]
    return indices_clases, clase_a_indice

def calcular_limites(clases):
    limite_inferior = min(clases)
    limite_superior = max(clases) + 1
    limites_inferiores = []
    limites_superiores = []
    while limite_inferior < limite_superior:
        limites_inferiores.append(limite_inferior)
        limites_superiores.append(limite_superior)
        limite_inferior += 1
        limite_superior += 1
    return limites_inferiores, limites_superiores

def calcular_marcas(limites_inferiores, limites_superiores):
    marcas_clase = [(li + ls) / 2 for li, ls in zip(limites_inferiores, limites_superiores)]
    return marcas_clase

def calcular_frecuencias(datos, clases):
    frecuencias_clase = [0] * len(clases)
    for dato in datos:
        indice_clase = clases.index(dato)
        frecuencias_clase[indice_clase] += 1
    return frecuencias_clase

def calcular_frecuencias_relativas(frecuencias_clase, total_datos):
    frecuencias_relativas = [f / total_datos for f in frecuencias_clase]
    return frecuencias_relativas

def calcular_frecuencias_acumuladas(frecuencias_clase):
    frecuencias_acumuladas = []
    acumulado = 0
    for frec in frecuencias_clase:
        acumulado += frec
        frecuencias_acumuladas.append(acumulado)
    return frecuencias_acumuladas

# Ejemplo de uso
datos = [3, 4, 5, 6, 7, 8, 3, 4, 5, 6]

# Calcular clases y asignar Ã­ndices
indices_clases, clase_a_indice = asignar_indices(datos)
clases = list(clase_a_indice.keys())

# Calcular lÃ­mites inferiores y superiores
limites_inferiores, limites_superiores = calcular_limites(clases)

# Calcular marcas de clase
marcas_clase = calcular_marcas(limites_inferiores, limites_superiores)

# Calcular frecuencias
frecuencias_clase = calcular_frecuencias(datos, clases)

# Calcular frecuencias relativas
total_datos = len(datos)
frecuencias_relativas = calcular_frecuencias_relativas(frecuencias_clase, total_datos)

# Calcular frecuencias acumuladas
frecuencias_acumuladas = calcular_frecuencias_acumuladas(frecuencias_clase)

# Formatear y mostrar los resultados en una tabla
print(f"{'Clase':<12} {'LÃ­mite Inferior':<16} {'LÃ­mite Superior':<16} {'Marca de Clase':<16} {'Frecuencia Absoluta':<22} {'Frecuencia Relativa':<22} {'Frecuencia Acumulada':<22}")
print("="*130)
for i in range(len(clases)):
    print(f"{clases[i]:<12} {limites_inferiores[i]:<16.1f} {limites_superiores[i]:<16.1f} {marcas_clase[i]:<16.1f} {frecuencias_clase[i]:<22} {frecuencias_relativas[i]:<22.2f} {frecuencias_acumuladas[i]:<22}")
```
